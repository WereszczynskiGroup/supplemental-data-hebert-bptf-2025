```python


# Ensure each df has a numeric residue ID column
for df in [df1, df2, df3, df4, df5]:
    df["ResID"] = df["Residue"].str.extract(r"(\d+)$").astype(int)

# Extract Total energy for residues 295–317 from each df
dfs = []
for i, df in enumerate([df1, df2, df3, df4, df5], start=1):
    subset = df[(df["ResID"] >= 295) & (df["ResID"] <= 317)].copy()
    subset = subset[["Residue", "Total"]].rename(columns={"Total": f"Total_{i}"})
    dfs.append(subset)

# Merge on Residue
merged = dfs[0]
for df in dfs[1:]:
    merged = pd.merge(merged, df, on="Residue")

# Calculate mean and std across the 5 Total columns
merged["Total_mean"] = merged[[f"Total_{i}" for i in range(1, 6)]].mean(axis=1)
merged["Total_std"] = merged[[f"Total_{i}" for i in range(1, 6)]].std(axis=1)

# After merging the Total columns:

# Also merge Total_err columns from each df
for i, df in enumerate([df1, df2, df3, df4, df5], start=1):
    subset_err = df[(df["ResID"] >= 295) & (df["ResID"] <= 317)].copy()
    subset_err = subset_err[["Residue", "Total_err"]].rename(columns={"Total_err": f"Total_err_{i}"})
    merged = pd.merge(merged, subset_err, on="Residue")

# Calculate mean of Total values (same as before)
merged["Total_mean"] = merged[[f"Total_{i}" for i in range(1, 6)]].mean(axis=1)

# Propagate the standard deviations of the Total_err columns
# Standard error of the mean = sqrt(sum of variances) / N
merged["Total_err_propagated"] = np.sqrt(
    np.sum(np.square(merged[[f"Total_err_{i}" for i in range(1, 6)]]), axis=1)
) / 5

# Final result with propagated error
result = merged[["Residue", "Total_mean", "Total_err_propagated"]]

print(result)
```
