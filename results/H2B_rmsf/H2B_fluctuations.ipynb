```python
#H2b1

import sys
import math 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 

yy,yyerr = np.loadtxt("Mh2b1.dat", usecols=(1,2), unpack=True, comments=('#','@','&'))
f,yy1,yy1err = np.loadtxt("EnoBh2b1.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&'))
g,yy2,yy2err = np.loadtxt("Eh2b1.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&'))


fig,ax=plt.subplots(figsize=(6.5,3.4))
plt.plot(f, yy, color = 'black',label = "NCP")
plt.plot(f, yy1, color = 'pink', label = "PHD + NCP")
plt.plot(f, yy2, color = 'olive', label = "BPTF + NCP")

plt.fill_between(f, yy-(yyerr/2.2), yy+(yyerr/2.2),color= 'black', antialiased='True', alpha=0.3)
plt.fill_between(f, yy1-(yy1err/2.2), yy1+(yy1err/2.2),color ='pink', antialiased='True', alpha=0.3)
plt.fill_between(f, yy2-(yy2err/2.2), yy2+(yy2err/2.2),color ='olive', antialiased='True', alpha=0.3)
ax.set_ylim(0,17)
ax.set_xlim(0,123)
ax.set_xlabel("Residue",fontsize=10)
ax.set_ylabel(r"RMSF ($\mathrm{\AA}$)",fontsize=10)
ax.legend(loc='center',labelspacing=0.1,borderpad=0.1)
plt.title("Nucleosome H2B.2 Fluctuations",fontsize=12)
plt.tight_layout()
plt.savefig('h2b2.pdf')
```
