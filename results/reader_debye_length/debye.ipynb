```python
import numpy as np
import matplotlib.pyplot as plt

x, y1 = np.loadtxt('./debye/bptf_debye1.dat', usecols=(0,1), unpack=True, comments=('#'))
y2 = np.loadtxt('./debye/bptf_debye2.dat', usecols=(1), unpack=True, comments=('#'))
y3 = np.loadtxt('./debye/bptf_debye3.dat', usecols=(1), unpack=True, comments=('#'))
y4 = np.loadtxt('./debye/bptf_debye4.dat', usecols=(1), unpack=True, comments=('#'))
y5 = np.loadtxt('./debye/bptf_debye5.dat', usecols=(1), unpack=True, comments=('#'))
y = np.concatenate((y1[-7000::], y2[-7000::], y3[-7000::], y4[-7000::], y5[-7000::]))

yy1 = np.loadtxt('./debye/phd_debye1.dat', usecols=(1), unpack=True, comments=('#'))
yy2 = np.loadtxt('./debye/phd_debye2.dat', usecols=(1), unpack=True, comments=('#'))
yy3 = np.loadtxt('./debye/phd_debye3.dat', usecols=(1), unpack=True, comments=('#'))
yy4 = np.loadtxt('./debye/phd_debye4.dat', usecols=(1), unpack=True, comments=('#'))
yy5 = np.loadtxt('./debye/phd_debye5.dat', usecols=(1), unpack=True, comments=('#'))
yy = np.concatenate((yy1[-7000::], yy2[-7000::], yy3[-7000::], yy4[-7000::], yy5[-7000::]))
```

```python
fig, axs = plt.subplots(5,2, sharey=True, sharex=True)

axs11 = axs[0,0].plot(x/100,yy1, color='black')
axs[0,0].set_xlim(0,1000)
axs[0,0].set_title('PHD + NCP')
axs12=axs[1,0].plot(x/100,yy2, color='black')
axs13=axs[2,0].plot(x/100,yy3, color='black')
axs14=axs[3,0].plot(x/100,yy4, color='black')
axs[2,0].set_ylabel('Atom Count')
axs15=axs[4,0].plot(x/100,yy5, color='black')
axs[4,0].set_xlabel('Time (ns)')

axs21 = axs[0,1].plot(x/100,y1, color='black')
axs[0,1].set_title('BPTF + NCP')
axs22 = axs[1,1].plot(x/100,y2, color='black')
axs23 = axs[2,1].plot(x/100,y3, color='black')
axs24 = axs[3,1].plot(x/100,y4, color='black')
axs25 = axs[4,1].plot(x/100,y5, color='black')

plt.suptitle('Reader Atoms within DNA Debye Length vs. Time')
ax0=axs[0,1].twinx()
ax0.set_yticks([])
ax0.set_ylabel('Run 1', rotation=315,  labelpad=25)
ax1=axs[1,1].twinx()
ax1.set_yticks([])
ax1.set_ylabel('Run 2', rotation=315,  labelpad=25)
ax2=axs[2,1].twinx()
ax2.set_yticks([])
ax2.set_ylabel('Run 3', rotation=315,  labelpad=25)
ax3=axs[3,1].twinx()
ax3.set_yticks([])
ax3.set_ylabel('Run 4', rotation=315,  labelpad=25)
ax4=axs[4,1].twinx()
ax4.set_yticks([])
ax4.set_ylabel('Run 5', rotation=315,  labelpad=25)

plt.tight_layout()
plt.savefig('debye_time.pdf', dpi=300)
```
