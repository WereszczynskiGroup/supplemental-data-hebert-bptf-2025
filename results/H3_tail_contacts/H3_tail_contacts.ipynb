```python
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
```

```python
# Iteratively loads data

Mdnadict = {}
Ednadict = {}
Ebptfdict = {}
EnoBdnadict = {}
EnoBphddict = {}
phd_dict = {}
bptf_dict = {}

for i in np.arange(0,42):
    Mdnadict[str(i)] = np.loadtxt('./H3_dna_contacts_nucleosome/'+str(999+i)+'_all.dat', unpack=True, usecols=(1), comments=('#'))
    EnoBdnadict[str(i)] = np.loadtxt('./H3_dna_contacts_phd/'+str(999+i)+'_all.dat', unpack=True, usecols=(1), comments=('#'))
    Ednadict[str(i)] = np.loadtxt('./H3_dna_contacts_bptf/'+str(999+i)+'_all.dat', unpack=True, usecols=(1), comments=('#'))
    EnoBphddict[str(i)] = np.loadtxt('./H3_reader_contacts_phd/'+str(999+i)+'_all.dat', unpack=True, usecols=(1), comments=('#'))
    Ebptfdict[str(i)] = np.loadtxt('./H3_reader_contacts_bptf/'+str(999+i)+'_all.dat', unpack=True, usecols=(1), comments=('#'))
for i in np.arange(0,15):
    phd_dict[str(i)] = np.loadtxt('./H3_1-15_phd_contacts/'+str(63+i)+'_all.dat', unpack=True, usecols=(1), comments=('#'))
    bptf_dict[str(i)] = np.loadtxt('./h3_1-15_bptf_contacts/'+str(171+i)+'_all.dat', unpack=True, usecols=(1), comments=('#'))

Mall = np.sum(Mdnadict[str(i)] for i in np.arange(0,15))
Ednall = np.sum(Ednadict[str(i)] for i in np.arange(0,15))
Ebptfall = np.sum(Ebptfdict[str(i)] for i in np.arange(0,15))
EnoBdnall = np.sum(EnoBdnadict[str(i)] for i in np.arange(0,15))
EnoBphdall = np.sum(EnoBphddict[str(i)] for i in np.arange(0,15))
phdall = np.sum(phd_dict[str(i)] for i in np.arange(0,15))
bptfall = np.sum(bptf_dict[str(i)] for i in np.arange(0,15))

Eall = Ednall+Ebptfall
EnoBall = EnoBdnall+EnoBphdall
```

```python
# This cell calculates the means and errors for use in the tables for the data table

print('NCP has '+str(Mall.mean())+'+/-'+str(stat.stdev(Mall)/2.2)+' tail-NCP contacts.')
print('PHD has '+str(phdall.mean())+'+/-'+str(stat.stdev(phdall)/2.2)+' tail-reader contacts.')
print('BPTF has '+str(bptfall.mean())+'+/-'+str(stat.stdev(bptfall)/2.2)+' tail-reader contacts.')
print('NCP+PHD has '+str(EnoBdnall.mean())+'+/-'+str(stat.stdev(EnoBdnall)/2.2)+' tail-NCP contacts.')
print('NCP+PHD has '+str(EnoBphdall.mean())+'+/-'+str(stat.stdev(EnoBphdall)/2.2)+' tail-reader contacts.')
print('NCP+PHD has '+str(EnoBall.mean())+'+/-'+str(stat.stdev(EnoBall)/2.2)+' total contacts.')
print('NCP+BPTF has '+str(Ednall.mean())+'+/-'+str(stat.stdev(Ednall)/2.2)+' tail-NCP contacts.')
print('NCP+BPTF has '+str(Ebptfall.mean())+'+/-'+str(stat.stdev(Ebptfall)/2.2)+' tail-reader contacts.')
print('NCP+BPTF has '+str(Eall.mean())+'+/-'+str(stat.stdev(Eall)/2.2)+' total contacts.')
```

```python
# Data table values are above cell's results, rounded to the nearest whole number for significant figures

data= [[ '94 $\pm$ 24', '-',  '94 $\pm$ 24'],
        [ '-', '200 $\pm$ 13',  '200 $\pm$ 13'],
        [ '61 $\pm$ 19', '191 $\pm$ 26', '252 $\pm$ 32'],
        [ '-',  '200 $\pm$ 12', '200 $\pm$ 12'],
        ['24 $\pm$ 18', '186 $\pm$ 15', '210 $\pm$ 22']]
```

```python
# Sorts data into arrays of average values
# Results should be arrays where every Nth element corresponds to the Nth residue

M = []
Msterr = []
Ed = []
Edsterr = []
EnoBd = []
EnoBdsterr = []
EnoBp = []
EnoBpsterr = []
Eb = []
Ebsterr = []
phd = []
phdsterr = []
bptf = []
bptfsterr = []

for i in np.arange(0,15):
    M.append(stat.mean(Mdnadict[str(i)]))
    Msterr.append(stat.stdev(Mdnadict[str(i)])/2.2)
    Ed.append(stat.mean(Ednadict[str(i)]))
    Edsterr.append(stat.stdev(Ednadict[str(i)])/2.2)
    EnoBd.append(stat.mean(EnoBdnadict[str(i)]))
    EnoBdsterr.append(stat.stdev(EnoBdnadict[str(i)])/2.2)
    Eb.append(stat.mean(Ebptfdict[str(i)]))
    Ebsterr.append(stat.stdev(Ebptfdict[str(i)])/2.2)
    EnoBp.append(stat.mean(EnoBphddict[str(i)]))
    EnoBpsterr.append(stat.stdev(EnoBphddict[str(i)])/2.2)
    phd.append(stat.mean(phd_dict[str(i)]))
    phdsterr.append(stat.stdev(phd_dict[str(i)])/2.2)
    bptf.append(stat.mean(bptf_dict[str(i)]))
    bptfsterr.append(stat.stdev(bptf_dict[str(i)])/2.2)
```

```python
# Creates Figure X
font = {'family': 'sans-serif',
        'color': 'black',
        'weight': 'bold',
        'size': 10,
        }

label = ['A1', 'R2', 'T3', 'K4me3', 'Q5', 'T6', 'A7', 'R8', 'K9', 'S10', 'T11', 'G12', 'G13', 'K14', 'A15']
l1 = np.arange(1, 16, 1)

width = 0.25

columns = ('DNA Contacts', 'Reader Contacts', 'Total')
rows = ('NCP', 'PHD + Peptide', 'NCP + PHD', 'BPTF + Peptide', 'NCP + BPTF')


fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(4.8,6.9))

# Plot 1
ax1.bar(l1 - width, height=np.array(M), yerr=np.array(Msterr), width=width, label='NCP', color='black', alpha=0.5, capsize=1.5)
ax1.bar(l1, np.array(Ed), yerr=np.array(Edsterr), width=width, label='NCP+BPTF', color='cyan', alpha=0.5, capsize=1.5)
ax1.bar(l1 + width, np.array(EnoBd), yerr=np.array(EnoBdsterr), width=width, label='NCP+PHD', color='pink', alpha=0.5, capsize=1.5)
ax1.set_title('H3k4me3 Tail-DNA Contacts', fontdict=font)
ax1.set_ylim(0, 75)
ax1.set_xticks(ticks=l1, labels=label, rotation='vertical', fontsize=10, fontweight='bold')

# Plot 2
ax2.bar(l1 - width * 0.5, np.array(Eb), yerr=np.array(Ebsterr), width=width, color='cyan', alpha=0.5, capsize=1.5)
ax2.bar(l1 + width * 0.5, np.array(EnoBp), yerr=np.array(EnoBpsterr), width=width, color='pink', alpha=0.5, capsize=1.5)
ax2.set_ylabel('Contacts', fontdict=font)
ax2.set_title('Tail-Reader Contacts in Presence of NCP', fontdict=font)
ax2.set_ylim(0, 75)
ax2.set_xticks(ticks=l1, labels=label, rotation='vertical', fontsize=10, fontweight='bold')

# Move legend from ax1 to ax2
handles, labels = ax1.get_legend_handles_labels()
ax2.legend(handles, labels, fontsize=10, loc='upper right')

# Plot 3
ax3.bar(l1 - width, np.array(phd), yerr=np.array(phdsterr), width=width, label='PHD', color='red', alpha=0.5, capsize=1.5)
ax3.bar(l1 + width * 0.5, np.array(bptf), yerr=np.array(bptfsterr), width=width, color='green', label='BPTF', alpha=0.5, capsize=1.5)
ax3.set_title('Tail-Reader Contacts, No NCP', fontdict=font)
ax3.set_ylim(0, 75)
ax3.set_xticks(ticks=l1, labels=label, rotation='vertical', fontsize=10, fontweight='bold')
ax3.legend(fontsize=10)

# Table
ax4.axis('off')  # Hide axis for the table
table = ax4.table(cellText=data, rowLabels=rows, colLabels=columns, cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1, 1.5)  # Adjust table size if needed

plt.tight_layout()
plt.savefig('h3tail_contacts.pdf')
```
