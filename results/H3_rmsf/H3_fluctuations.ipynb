```python
import math 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 

y,yerr = np.loadtxt("Nucm_h3_1.dat", usecols=(1,2), unpack=True, comments=('#','@','&'))
f,y1,y1err = np.loadtxt("EnoB_h3_1.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&'))
g,y2,y2err = np.loadtxt("NucE_h3_1.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&'))
x = np.array([i for i in range(0,135)])


fig,ax=plt.subplots(figsize=(7,3.4))
plt.plot(x, y, color = 'black',label = "NCP")
plt.plot(x, y1, color = 'pink', label = "PHD + NCP")
plt.plot(x, y2, color = 'olive', label = "BPTF + NCP")

plt.fill_between(x, y-(yerr/2.2), y+(yerr/2.2),color= 'black', antialiased='True', alpha=0.3)
plt.fill_between(x, y1-(y1err/2.2), y1+(y1err/2.2),color ='pink', antialiased='True', alpha=0.3)
plt.fill_between(x, y2-(y2err/2.2), y2+(y2err/2.2),color ='olive', antialiased='True', alpha=0.3)
ax.set_ylim(0,24)
ax.set_xlim(0,45)
plt.xticks(np.arange(45),['A1','R2','T3','K4me3','Q5','T6','A7','R8','K9','S10','T11','G12','G13','K14','A15','P16','R17','K18','Q19','L20','A21','T22','K23','A24','A25','R26','K27','S28','A29','P30','A31','T32','G33','G34','V35','K36','K37','P38','H39','R40','Y41','R42','P43', 'G44', 'T45'],rotation='vertical',fontsize=10)
ax.set_yticklabels(np.arange(0,26,2), fontsize=8)
plt.yticks(np.arange(0, 26, 2.0) )
ax.set_ylabel(r"RMSF ($\mathrm{\AA}$)",fontsize=10)
plt.axvline(x=36.0, color='k', linestyle='--')
bbox_props = dict(boxstyle="larrow,pad=0.3", fc="papayawhip", ec="k", lw=2)
t = ax.text(33, 10, "Tail ", ha="center", va="center",size=12,bbox=bbox_props)
bbox_props = dict(boxstyle="rarrow,pad=0.3", fc="papayawhip", ec="k", lw=2)
t = ax.text(39.5, 10, "Core", ha="center", va="center",size=12,bbox=bbox_props)
ax.legend(loc='upper right',labelspacing=0.1,borderpad=0.1)
plt.title("Nucleosome H3K4me3 1-45 Fluctuations",fontsize=12)
plt.savefig('h3k4me3_fluct.pdf')
```
