```python
import matplotlib.pyplot as plt
import numpy as np


x1, y1 = np.loadtxt("bptf_angle/bptf_angle1.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))
x2,y2 = np.loadtxt("bptf_angle/bptf_angle2.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))
x3,y3 = np.loadtxt("bptf_angle/bptf_angle3.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))
x4,y4 = np.loadtxt("bptf_angle/bptf_angle4.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))
x5,y5 = np.loadtxt("bptf_angle/bptf_angle5.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))
y0 = np.concatenate((y1[-7000:],y2[-7000:],y3[-7000:],y4[-7000:],y5[-7000:]))
y = np.concatenate((y1,y2,y3,y4,y5))
x = np.concatenate((x1,x2,x3,x4,x5))

f1 = np.loadtxt("bptf_angle/bptf_peptide_angle1.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f2 = np.loadtxt("bptf_angle/bptf_peptide_angle2.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f3 = np.loadtxt("bptf_angle/bptf_peptide_angle3.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f4 = np.loadtxt("bptf_angle/bptf_peptide_angle4.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f5 = np.loadtxt("bptf_angle/bptf_peptide_angle5.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f0 = np.concatenate((f1[-7000:],f2[-7000:],f3[-7000:],f4[-7000:],f5[-7000:]))
f = np.concatenate((f1,f2,f3,f4,f5))

d1 = np.loadtxt("bptf_angle/bptf_ncp_angle1.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d2 = np.loadtxt("bptf_angle/bptf_ncp_angle2.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d3 = np.loadtxt("bptf_angle/bptf_ncp_angle3.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d4 = np.loadtxt("bptf_angle/bptf_ncp_angle4.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d5 = np.loadtxt("bptf_angle/bptf_ncp_angle5.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d0 = np.concatenate((d1[-7000:],d2[-7000:],d3[-7000:],d4[-7000:],d5[-7000:]))
d = np.concatenate((d1,d2,d3,d4,d5))


fig, axs = plt.subplots(5,3, sharex=True, sharey=True)
fig.suptitle('BPTF Angle Between Domains vs. Time', fontweight='bold', fontsize=12)
axs[0,0].plot(x1/10, y1, color='black')
axs[0,0].set_title('BPTF')
axs[0,0].set_xlim(0,1000)
axs[0,0].set_yticks(np.arange(0,240,60))
axs[0,0].set_yticklabels(('0','60','120','180'))
axs[1,0].plot(x2/10, y2, color='black')
axs[2,0].plot(x3/10, y3, color='black')
axs[2,0].set_ylabel('PHD - Bromodomain Angle (degrees)', fontsize=12)
axs[3,0].plot(x4/10, y4, color='black')
axs[4,0].plot(x5/10, y5, color='black')
axs[4,0].set_xticks(np.arange(0,1250,250))
axs[4,0].set_xticklabels(('0','','500','','1000'))

xmin, xmax = axs[0,1].get_xlim()
text_y = axs[0,1].get_ylim()[1] * 1.50
axs[0,1].plot(x1/10, f1, color='black')
axs[0,1].set_title('BPTF + Peptide')
axs[1,1].plot(x2/10, f2, color='black')
axs[2,1].plot(x3/10, f3, color='black')
axs[3,1].plot(x4/10, f4, color='black')
axs[4,1].plot(x5/10, f5, color='black')
axs[4,1].set_xlabel('Time (ns)')
axs[0,2].plot(x1/10, d1, color='black')
axs[0,2].set_title('BPTF + NCP')
ax0=axs[0,2].twinx()
ax0.set_yticks([])
ax0.set_ylabel('Run 1', rotation=315,  labelpad=25)
axs[1,2].plot(x2/10, d2, color='black')
ax1=axs[1,2].twinx()
ax1.set_yticks([])
ax1.set_ylabel('Run 2', rotation=315,  labelpad=25)
axs[2,2].plot(x3/10, d3, color='black')
ax2=axs[2,2].twinx()
ax2.set_yticks([])
ax2.set_ylabel('Run 3', rotation=315,  labelpad=25)
axs[3,2].plot(x4/10, d4, color='black')
ax3=axs[3,2].twinx()
ax3.set_yticks([])
ax3.set_ylabel('Run 4', rotation=315,  labelpad=25)
axs[4,2].plot(x5/10, d5, color='black')
ax4=axs[4,2].twinx()
ax4.set_yticks([])
ax4.set_ylabel('Run 5', rotation=315,  labelpad=25)

plt.savefig('angles_time.pdf')
plt.tight_layout()
plt.show()
```
