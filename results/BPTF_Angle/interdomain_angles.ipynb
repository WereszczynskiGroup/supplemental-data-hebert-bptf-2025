```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec
from PIL import Image

yy1 = np.loadtxt("Notailangle1.dat", usecols=(1), unpack=True, comments=('#','@','&'))
yy2 = np.loadtxt("Notailangle2.dat", usecols=(1), unpack=True, comments=('#','@','&'))
yy3 = np.loadtxt("Notailangle3.dat", usecols=(1), unpack=True, comments=('#','@','&'))
yy4 = np.loadtxt("Notailangle4.dat", usecols=(1), unpack=True, comments=('#','@','&'))
yy5 = np.loadtxt("Notailangle5.dat", usecols=(1), unpack=True, comments=('#','@','&'))
yy0 = np.concatenate((yy1[-70000:],yy2[-70000:],yy3[-70000:],yy4[-70000:],yy5[-70000:]))

dd1 = np.loadtxt("2f6jexangle1.dat", usecols=(1), unpack=True, comments=('#','@','&'))
dd2 = np.loadtxt("2f6jexangle2.dat", usecols=(1), unpack=True, comments=('#','@','&'))
dd3 = np.loadtxt("2f6jexangle3.dat", usecols=(1), unpack=True, comments=('#','@','&'))
dd4 = np.loadtxt("2f6jexangle4.dat", usecols=(1), unpack=True, comments=('#','@','&'))
dd5 = np.loadtxt("2f6jexangle5.dat", usecols=(1), unpack=True, comments=('#','@','&'))
dd0 = np.concatenate((dd1[-70000:],dd2[-70000:],dd3[-70000:],dd4[-70000:],dd5[-70000:]))

ff1 = np.loadtxt("Eangle1.dat", usecols=(1), unpack=True, comments=('#','@','&'))
ff2 = np.loadtxt("Eangle2.dat", usecols=(1), unpack=True, comments=('#','@','&'))
ff3 = np.loadtxt("Eangle3.dat", usecols=(1), unpack=True, comments=('#','@','&'))
ff4 = np.loadtxt("Eangle4.dat", usecols=(1), unpack=True, comments=('#','@','&'))
ff5 = np.loadtxt("Eangle5.dat", usecols=(1), unpack=True, comments=('#','@','&'))
ff0 = np.concatenate((ff1[-70000:],ff2[-70000:],ff3[-70000:],ff4[-70000:],ff5[-70000:]))

im1 = Image.open('rg17.5.png')
im2 = Image.open('rg29.1.png')
fig = plt.figure(figsize=(5, 3), dpi=300)
gs = GridSpec(2,3)

ax1 = fig.add_subplot(gs[:,:2])

ax1.hist(yy0, alpha=0.5, linewidth=2, density=True, bins=300, histtype='step', color='purple', label='Free')
ax1.hist(dd0, alpha=0.5, linewidth=2, density=True, bins=300, histtype='step', color='orange', label='+Peptide')
ax1.hist(ff0, alpha=0.5, bins=300, linewidth=2, density=True, histtype='step', color='olive', label='+NCP')
ax1.set_title("BPTF Finger-linker-bromodomain Angle")
ax1.set_xlabel("Angle (degrees)")
ax1.set_ylabel("Probability Density")
ax1.legend(loc='upper left', fontsize=8)
ax2 = fig.add_subplot(gs[0,2])
ax2.imshow(im1)
ax2.axis('off')  # clear axes from graphic
ax2.set_title('37.3$^\circ$', fontsize=8, y=-0.2)

ax4 = fig.add_subplot(gs[1,2])
ax4.imshow(im2)
ax4.axis('off')  # clear axes from graphic
ax4.set_title('133.7$^\circ$', fontsize=8, y=-0.2)
plt.tight_layout()
plt.savefig("Angles.pdf")
```
