```python
import numpy as np
import statistics as stat

# Note: In my systems, indices for the start of the different data sets are 0, 700, and 1400 for the complex, receptor, and ligand respectively

time1 = np.loadtxt(M1_1, unpack=True,comments=('#'),delimiter=',',usecols=0)
vdwaals1 = np.loadtxt(M1_1, unpack=True,comments=('#'),delimiter=',',usecols=1)
eel1 = np.loadtxt(M1_1, unpack=True,comments=('#'),delimiter=',',usecols=2)
egb1 = np.loadtxt(M1_1, unpack=True,comments=('#'),delimiter=',',usecols=3)
esurf1 = np.loadtxt(M1_1, unpack=True,comments=('#'),delimiter=',',usecols=4)
total1 = np.loadtxt(M1_1, unpack=True,comments=('#'),delimiter=',',usecols=7)

time2 = np.loadtxt(M1_2, unpack=True,comments=('#'),delimiter=',',usecols=0)
vdwaals2 = np.loadtxt(M1_2, unpack=True,comments=('#'),delimiter=',',usecols=1)
eel2 = np.loadtxt(M1_2, unpack=True,comments=('#'),delimiter=',',usecols=2)
egb2 = np.loadtxt(M1_2, unpack=True,comments=('#'),delimiter=',',usecols=3)
esurf2 = np.loadtxt(M1_2, unpack=True,comments=('#'),delimiter=',',usecols=4)
total2 = np.loadtxt(M1_2, unpack=True,comments=('#'),delimiter=',',usecols=7)

time3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=0)
bond3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=1)
angle3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=2)
dihed3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=3)
vdwaals3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=7)
eel3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=8)
vdw_1_4_3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=9)
eel_1_4_3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=10)
egb3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=11)
esurf3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=12)
total3 = np.loadtxt(M1_3, unpack=True,comments=('#'),delimiter=',',usecols=15)

time4 = np.loadtxt(M1_4, unpack=True,comments=('#'),delimiter=',',usecols=0)
vdwaals4 = np.loadtxt(M1_4, unpack=True,comments=('#'),delimiter=',',usecols=1)
eel4 = np.loadtxt(M1_4, unpack=True,comments=('#'),delimiter=',',usecols=2)
egb4 = np.loadtxt(M1_4, unpack=True,comments=('#'),delimiter=',',usecols=3)
esurf4 = np.loadtxt(M1_4, unpack=True,comments=('#'),delimiter=',',usecols=4)
total4 = np.loadtxt(M1_4, unpack=True,comments=('#'),delimiter=',',usecols=7)

time5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=0)
bond5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=1)
angle5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=2)
dihed5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=3)
vdwaals5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=7)
eel5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=8)
vdw_1_4_5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=9)
eel_1_4_5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=10)
egb5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=11)
esurf5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=12)
total5 = np.loadtxt(M1_5, unpack=True,comments=('#'),delimiter=',',usecols=15)


######### These values should be generalized to work with any .xvg output - no editing required! ##########
### Comp0:compf covers complex, rec0:recf receptor, and lig0:ligf ligand ###
comp0 = 0
compf = int(len(time1)/4)
rec0 = compf 
recf = int(2*len(time1)/4)
lig0 = recf
ligf = int(3*len(time1)/4)

comp2 = int(len(time3)/3)
rec20 = comp2
rec2f = int(2*len(time3)/3)
lig20 = rec2f
lig2f = int(len(time3))
###########################################################################################################

Elec1 = eel1 + egb1 
VdW1 = vdwaals1 + esurf1

Elec2 = eel2 + egb2
VdW2 = vdwaals2 + esurf2

Elec3 = eel3 + egb3
VdW3 = vdwaals3 + esurf3

Elec4 = eel4 + egb4
VdW4 = vdwaals4 + esurf4

Elec5 = eel5 + egb5
VdW5 = vdwaals5 + esurf5

######
## Concatenate everything into longer time series ### 
VdWcom = np.concatenate((VdW1[comp0:compf], VdW2[comp0:compf], VdW3[comp0:comp2], VdW4[comp0:compf], VdW5[comp0:comp2]))
VdWrec = np.concatenate((VdW1[rec0:recf], VdW2[rec0:recf], VdW3[rec20:rec2f],VdW4[rec0:recf], VdW5[rec20:rec2f]))
VdWlig = np.concatenate((VdW1[lig0:ligf], VdW2[lig0:ligf], VdW3[lig20:lig2f], VdW4[lig0:ligf], VdW5[lig20:lig2f]))

Eleccom = np.concatenate((Elec1[comp0:compf], Elec2[comp0:compf], Elec3[comp0:comp2], Elec4[comp0:compf], Elec5[comp0:comp2]))
Elecrec = np.concatenate((Elec1[rec0:recf], Elec2[rec0:recf], Elec3[rec20:rec2f], Elec4[rec0:recf], Elec5[rec20:rec2f]))
Eleclig = np.concatenate((Elec1[lig0:ligf], Elec2[lig0:ligf], Elec3[lig20:lig2f], Elec4[lig0:ligf], Elec5[lig20:lig2f]))

totalcom = Eleccom + VdWcom
######


with open(output_file1, "w") as f:
    f.write("# Complex\n # Elec VdW\n")  # Header
    for i in range(len(VdWcom)):
        f.write(f"{i} {Eleccom[i]:.6f} {VdWcom[i]:.6f}\n")
print(f"Data written to {output_file1}")


with open(output_file2, "w") as f:
    f.write("# Receptor\n # Elec VdW\n")  # Header
    for i in range(len(VdWrec)):
        f.write(f"{i} {Elecrec[i]:.6f} {VdWrec[i]:.6f}\n")
print(f"Data written to {output_file2}")


with open(output_file3, "w") as f:
    f.write("# Ligand\n # Elec VdW\n")  # Header
    for i in range(len(VdWlig)):
        f.write(f"{i} {Eleclig[i]:.6f} {VdWlig[i]:.6f}\n")
print(f"Data written to {output_file3}")
```

```python
# Define sum of squares function for propagating error
def sumsq(a,b,c):
    sumsquares = a*a + b*b + c*c
    return sumsquares

###############
SI = timeseries.statistical_inefficiency(totalcom)
N = np.floor(len(totalcom)/SI)
print('This inefficiency test is based on the time series of the complex total energy.')
print('The number of significant points is '+str(N)+'. Decorrelation time = '+str(len(totalcom)/N)+' frames \n')

avgVdW = np.average(VdWcom) - np.average(VdWrec) - np.average(VdWlig)

avgEl = np.average(Eleccom) - np.average(Elecrec) - np.average(Eleclig)

avgtotal = avgEl + avgVdW

VdWsterr = sqrt(sumsq(np.std(VdWcom)/sqrt(N),np.std(VdWlig)/sqrt(N),np.std(VdWrec)/sqrt(N))) 
r_avgVdW = np.round(avgVdW, decimals=1)
r_VdWsterr = np.round(VdWsterr, decimals=1)


Elsterr = sqrt(sumsq(np.std(Eleccom)/sqrt(N),np.std(Eleclig)/sqrt(N),np.std(Elecrec)/sqrt(N))) 
r_avgEl = np.round(avgEl, decimals=1)
r_Elsterr = np.round(Elsterr, decimals=1)


totalsterr = sqrt(sumsq(VdWsterr,Elsterr,0))
r_avgtotal = np.round(avgtotal,decimals=1)
r_totalsterr = np.round(totalsterr,decimals=1)

print('VdW: '+str(r_avgVdW)+' +/- '+str(r_VdWsterr)+'\n')
print('Elec: '+str(r_avgEl)+' +/- '+str(r_Elsterr)+'\n')
print('Total: '+str(r_avgtotal)+' +/- '+str(r_totalsterr)+'\n')
```
