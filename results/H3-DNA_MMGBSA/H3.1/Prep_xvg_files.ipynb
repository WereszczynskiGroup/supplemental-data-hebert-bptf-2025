```python
# Run this on every .xvg file you intend to use
# It will comment out nonnumerical rows so the columns can be loaded into numpy vectors

import re
import sys
import numpy as np
import statistics as stat
from pymbar import timeseries
from math import sqrt

M1_1 = './BPTF_NCP/E1gbsa.xvg'
M1_2 = './BPTF_NCP/E2gbsa.xvg'
M1_3 = './BPTF_NCP/E3gbsa.xvg'
M1_4 = './BPTF_NCP/E4gbsa.xvg'
M1_5 = './BPTF_NCP/E5gbsa.xvg'

def edit_xvg_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    
    with open(filename, 'w') as file:
        for line in lines:
            # If the line contains non-numeric characters (excluding commas), prefix it with '#'
            if re.search(r'[^0-9.,\-\s]', line):
                file.write(f'#{line}')
            else:
                file.write(line)
    print(filename+' edited')

files = [M1_1,M1_2,M1_3,M1_4,M1_5]
for i in files:
    edit_xvg_file(i)
```
