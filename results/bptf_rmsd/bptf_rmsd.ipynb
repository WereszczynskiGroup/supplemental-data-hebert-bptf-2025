```python
x = x1/10
fig, axs = plt.subplots(3,3, sharey='row', sharex='col')

axs[0,0].plot(x,y1, alpha=0.5)
axs[0,0].plot(x,y2, alpha=0.5)
axs[0,0].plot(x,y3, alpha=0.5)
axs[0,0].plot(x,y4, alpha=0.5)
axs[0,0].plot(x,y5, alpha=0.5)
axs[0,0].set_title('Finger RMSD')

axs[0,1].plot(x,yy1, alpha=0.5)
axs[0,1].plot(x,yy2, alpha=0.5)
axs[0,1].plot(x,yy3, alpha=0.5)
axs[0,1].plot(x,yy4, alpha=0.5)
axs[0,1].plot(x,yy5, alpha=0.5)
axs[0,1].set_title('Bromodomain RMSD')

axs[0,2].plot(x,yyy1, alpha=0.5, label='Run 1')
axs[0,2].plot(x,yyy2, alpha=0.5, label='Run 2')
axs[0,2].plot(x,yyy3, alpha=0.5, label='Run 3')
axs[0,2].plot(x,yyy4, alpha=0.5, label='Run 4')
axs[0,2].plot(x,yyy5, alpha=0.5, label='Run 5')
axs[0,2].set_title('BPTF RMSD')

axs[1,0].plot(x,z1, alpha=0.5)
axs[1,0].plot(x,z2, alpha=0.5)
axs[1,0].plot(x,z3, alpha=0.5)
axs[1,0].plot(x,z4, alpha=0.5)
axs[1,0].plot(x,z5, alpha=0.5)
axs[1,0].set(ylabel='RMSD ($\AA$)')

axs[1,1].plot(x,zz1, alpha=0.5)
axs[1,1].plot(x,zz2, alpha=0.5)
axs[1,1].plot(x,zz3, alpha=0.5)
axs[1,1].plot(x,zz4, alpha=0.5)
axs[1,1].plot(x,zz5, alpha=0.5)

axs[1,2].plot(x,zzz1, alpha=0.5)
axs[1,2].plot(x,zzz2, alpha=0.5)
axs[1,2].plot(x,zzz3, alpha=0.5)
axs[1,2].plot(x,zzz4, alpha=0.5)
axs[1,2].plot(x,zzz5, alpha=0.5)

axs[2,0].plot(x,a1, alpha=0.5)
axs[2,0].plot(x,a2, alpha=0.5)
axs[2,0].plot(x,a3, alpha=0.5)
axs[2,0].plot(x,a4, alpha=0.5)
axs[2,0].plot(x,a5, alpha=0.5)

axs[2,1].set(xlabel='Time (ns)')

axs[2,1].plot(x,aa1, alpha=0.5)
axs[2,1].plot(x,aa2, alpha=0.5)
axs[2,1].plot(x,aa3, alpha=0.5)
axs[2,1].plot(x,aa4, alpha=0.5)
axs[2,1].plot(x,aa5, alpha=0.5)

axs[2,2].plot(x,aaa1, alpha=0.5)
axs[2,2].plot(x,aaa2, alpha=0.5)
axs[2,2].plot(x,aaa3, alpha=0.5)
axs[2,2].plot(x,aaa4, alpha=0.5)
axs[2,2].plot(x,aaa5, alpha=0.5)

for ax in axs.flat:
    ax.set_yticks([0, 5, 10, 15])
    ax.set_xticks(np.arange(0,1250,250))
    ax.set_xticklabels(('0', '', '500', '', '1000'))
    ax.set_xlim(0,1000)
    ax.set_ylim(0,17)

ax1 = axs[0,2].twinx()
ax1.set_yticks([])
ax1.set_ylabel('Solution BPTF', rotation=270, labelpad=15)
ax1 = axs[1,2].twinx()
ax1.set_yticks([])
ax1.set_ylabel('BPTF + Peptide', rotation=270, labelpad=15)
ax1 = axs[2,2].twinx()
ax1.set_yticks([])
ax1.set_ylabel('BPTF + NCP', rotation=270, labelpad=15)
legend = axs[0, 2].legend(loc='upper left', bbox_to_anchor=(1.15, 1))

plt.tight_layout() 
plt.savefig('RMSD_comparison.pdf', dpi=150)
```
