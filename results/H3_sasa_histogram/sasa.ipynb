```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
font = {'family': 'sans-serif',
        'color':  'black',
        'weight': 'normal',
        'size': 12,
        }

x = np.loadtxt("nucleosomeH3SASAall.dat", usecols=(1), unpack=True, comments=('#','@','&'))
z = np.loadtxt("bptf_ncpH3SASAall.dat", usecols=(1), unpack=True, comments=('#','@','&'))
b = np.loadtxt("phd_ncpH3SASAall.dat", usecols=(1), unpack=True, comments=('#','@','&'))

plt.hist(x, alpha=0.5, bins=100, density=True, linewidth=3, label='NCP', histtype='step')

plt.hist(b, alpha=0.5, bins=100, density=True, linewidth=3, label='PHD + NCP', histtype='step')

plt.hist(z, alpha=0.5, bins=100, density=True, linewidth=3, label='BPTF + NCP', histtype='step')


plt.xlabel('Solvent Accessible Surface Area ($\AA^2$)', fontdict=font)
plt.ylabel('Frequency', fontdict=font)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('H3K4me3 Solvent Accessible Surface Area', fontdict=font, fontweight='bold')
plt.legend(prop={'size': 12}, loc='upper right')
plt.tight_layout()
plt.savefig('Tail_SASA.pdf')
plt.show()
```
